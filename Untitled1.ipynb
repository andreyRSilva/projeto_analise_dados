# 4. Analisando o dataset

```python
df['DATA PEDIDO'] = pd.to_datetime(df['DATA PEDIDO'])
df['MÊS'] = df['DATA PEDIDO'].dt.month
lucro_por_mes = df.groupby('MÊS')['VALOR TOTAL'].sum().reset_index()
display(lucro_por_mes)


#criando o grafico 
#transformando a coluna de data para o formato datetime
df['DATA PEDIDO'] = pd.to_datetime(df['DATA PEDIDO'])

# Criando colunas de mês e número do mês
df['NUM_MES'] = df['DATA PEDIDO'].dt.month
df['MÊS'] = df['DATA PEDIDO'].dt.month_name()

# Agrupando por mês
lucro_por_mes = df.groupby(['NUM_MES','MÊS'])['VALOR TOTAL'].sum().reset_index()
lucro_por_mes = lucro_por_mes.sort_values('NUM_MES')

# Gráfico


plt.bar(lucro_por_mes['MÊS'], lucro_por_mes['VALOR TOTAL'], color='skyblue')
plt.title("Lucro por mês")
plt.xlabel("Mês")
plt.ylabel("Lucro (R$)")
plt.xticks(rotation=45)
plt.show()
```

Como podemos observar na tabela, existe um cresimento continuo durante o ano com um nivel marcante no mês de dedezembro sendo o mês mais lucrativo com 26,76% do lucro anual acompanhado pelo mes de novembro com 12,14%.

```python
media_por_mes = (
    df.groupby(['NUM_MES', 'MÊS'])['VALOR TOTAL']
      .mean()
      .reset_index()
      .rename(columns={'VALOR TOTAL': 'VALOR_MEDIO'})
      .sort_values('NUM_MES') 
)
display(media_por_mes)
#criando o grafico

plt.plot(media_por_mês["MÊS"], media_por_mês["VALOR MEDIO"], marker='o', linestyle='-')
plt.title("Média por mês")
plt.xlabel("mês")
plt.ylabel("Valor medio")
plt.xticks(rotation=45)  # girar nomes dos meses
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

O valor mensal se mantém relativamente estavel ao longo do ano.

```python
lucro_por_categoria = df.groupby('CATEGORIA')['VALOR TOTAL'].sum().reset_index()
lucro_por_categoria = lucro_por_categoria.sort_values(by='VALOR TOTAL', ascending=False)
display(lucro_por_categoria)
```

Podemos observar uma maior demanda por determinadas categorias de produtos, sendo interessante concentrar investimentos em 6 das 19 categorias, que juntas representam 80% do lucro anual. Essas categorias são: proteínas, aminoácidos, pré-treinos e energéticos, carboidratos, emagrecimento e vitaminas e minerais

```python
lucro_por_estado = df.groupby('ESTADO')['VALOR TOTAL'].sum().reset_index()
lucro_por_estado = lucro_por_estado.sort_values(by='VALOR TOTAL', ascending=False)

frete_por_estado = df.groupby('ESTADO')['FRETE'].sum().reset_index()

tabela_lado_a_lado = pd.merge(lucro_por_estado, frete_por_estado, on='ESTADO')
print(tabela_lado_a_lado)
```

Com este dataset, podemos visualizar o lucro de cada mês e seus respectivos custos de frete. Inicialmente, observamos que o Rio Grande do Sul apresenta
um dos menores valores de frete, enquanto Santa Catarina, que possui um dos menores valores em compras, apresenta um dos maiores custos de frete.
Também notamos que Sergipe registra o segundo maior lucro, porém com um custo de frete elevado.
algumas soluções possíveis incluem:
 *Negociar preços melhores com transportadoras regionais para reduzir o frete.
 *Evitar enviar produtos de baixa margem para estados com custos de frete altos.

```python
frete_por_fornecedor = df.groupby('FORNECEDOR')['FRETE'].sum().reset_index()
lucro_por_fornecedor = df.groupby('FORNECEDOR')['VALOR TOTAL'].sum().reset_index()
tabela_lado_a_lado = pd.merge(lucro_por_fornecedor, frete_por_fornecedor, on='FORNECEDOR')

tabela_lado_a_lado = tabela_lado_a_lado.sort_values(by='VALOR TOTAL', ascending=False)
print(tabela_lado_a_lado)
```

Podemos observar que os fornecedores 01, 02, 08 e 09 apresentam destaque positivo em relação aos custos de frete. Por outro lado, o fornecedor 06 apresenta um alto custo de frete, mesmo com menor lucro total, indicando que não é uma opção vantajosa de fornecedor.
